```python
import pandas as pd 
```

```python
T20_Matches = pd.read_csv('Pakistan_T20_Matches_Information.csv')
matches_info = pd.read_csv('match_info.csv')
players_name = pd.read_csv('Player_name.csv')
```

```python
#file cleaning process
players_name['name'] = players_name['name'].astype("category")
players_name['team'] = players_name['team'].astype("category")
players_name['match_type_number'] = players_name['match_type_number'].astype("int16")
players_name['match_type'] = players_name['match_type'].astype("category")
players_name['venue'] = players_name['venue'].astype("category")
```

```python
players_name.to_csv("T20_Players_Name.csv",index=False)
```

**MATHE INFO CSV CLEANING**

```python
matches_info.head()
```

```python
matches_info.info()
```

```python
#Converting to suitable types for memory efficiency
matches_info['match_type_number'] = matches_info['match_type_number'].astype('int16')
matches_info['overs'] = matches_info['overs'].astype('int16')
matches_info['city'] = matches_info['city'].astype('category')
matches_info['event'] = matches_info['event'].astype('category')
matches_info['stage'] = matches_info['stage'].astype('category')
matches_info['dates'] = matches_info['dates'].astype('category')
matches_info['match_type'] = matches_info['match_type'].astype('category')
matches_info['winner'] = matches_info['winner'].astype('category')
matches_info['season'] = matches_info['season'].astype('category')
matches_info['team_type'] = matches_info['team_type'].astype('category')
matches_info['team1'] = matches_info['team1'].astype('category')
matches_info['team2'] = matches_info['team2'].astype('category')
matches_info['toss_winner'] = matches_info['toss_winner'].astype('category')
matches_info['toss_decision'] = matches_info['toss_decision'].astype('category')
matches_info['venue'] = matches_info['venue'].astype('category')
```

```python
matches_info.to_csv("T20_Matches_Info.csv")
```

```python
T20_Matches.head()
```

```python
T20_Matches['Over'] = T20_Matches['Over'].astype('int16')
T20_Matches['Runs_Batter'] = T20_Matches['Runs_Batter'].astype('int16')
T20_Matches['Runs_Extras'] = T20_Matches['Runs_Extras'].astype('int16')
T20_Matches['Runs_Total'] = T20_Matches['Runs_Total'].astype('int16')
T20_Matches['Wickets'] = T20_Matches['Wickets'].astype('int16')
T20_Matches['Team'] = T20_Matches['Team'].astype('category')
T20_Matches['Batter'] = T20_Matches['Batter'].astype('category')
T20_Matches['Bowler'] = T20_Matches['Bowler'].astype('category')
T20_Matches['Non_Striker'] = T20_Matches['Non_Striker'].astype('category')
T20_Matches['fielder_name'] = T20_Matches['fielder_name'].fillna('Unknown')
T20_Matches['fielder_name'] = T20_Matches['fielder_name'].astype('category')
T20_Matches['kind'] = T20_Matches['kind'].fillna('Not Out')
T20_Matches['kind'] = T20_Matches['kind'].astype('category')
T20_Matches['player_out'] = T20_Matches['player_out'].fillna('Not Out')
T20_Matches['player_out'] = T20_Matches['player_out'].astype('category')
T20_Matches['match_type_number'] = pd.to_numeric(T20_Matches['match_type_number'], errors='coerce')
T20_Matches['match_type_number'] = T20_Matches['match_type_number'] .fillna(T20_Matches['match_type_number'].mean())
```

```python
T20_Matches.info()
```

```python
T20_Matches.to_csv("Pakistan_T20_Players_Innings_Infor.csv",index=False)
```
